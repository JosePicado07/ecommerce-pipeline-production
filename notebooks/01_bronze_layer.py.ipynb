### We begin the creation of the Full Bronze Layer

```python
from pyspark.sql.functions import current_timestamp, lit

customers_df = spark.table("olist_customers_dataset")
customers_bronze = customers_df \
  .withColumn("ingesttime", current_timestamp()) \
  .withColumn("data_source", lit("olist")) \
  .withColumn("data_layer", lit("bronze")) \
  .withColumn("data_status",lit("raw_unaltered"))

orders_df = spark.table("olist_orders_dataset")
orders_bronze = orders_df \
  .withColumn("ingesttime", current_timestamp()) \
  .withColumn("data_source", lit("olist")) \
  .withColumn("data_layer", lit("bronze")) \
  .withColumn("data_status",lit("raw_unaltered"))


order_items_df = spark.table("olist_order_items_dataset")
order_items_bronze = order_items_df \
  .withColumn("ingesttime", current_timestamp()) \
  .withColumn("data_source", lit("olist")) \
  .withColumn("data_layer", lit("bronze")) \
  .withColumn("data_status",lit("raw_unaltered"))

products_df = spark.table("olist_products_dataset")
products_bronze = products_df \
  .withColumn("ingesttime", current_timestamp()) \
  .withColumn("data_source", lit("olist")) \
  .withColumn("data_layer", lit("bronze")) \
  .withColumn("data_status", lit("raw_unaltered"))

spark.sql("CREATE DATABASE IF NOT EXISTS bronze")  

customers_bronze.write \
  .format("delta") \
  .mode("overwrite") \
  .saveAsTable("bronze.customers")

orders_bronze.write \
  .format("delta") \
  .mode("overwrite") \
  .saveAsTable("bronze.orders")

order_items_bronze.write \
  .format("delta") \
  .mode("overwrite") \
  .saveAsTable("bronze.order_items")

products_bronze.write \
  .format("delta") \
  .mode("overwrite") \
  .saveAsTable("bronze.products")


display(spark.sql("DESCRIBE DETAIL bronze.customers"))
display(spark.sql("DESCRIBE DETAIL bronze.orders"))
display(spark.sql("DESCRIBE DETAIL bronze.order_items"))
display(spark.sql("DESCRIBE DETAIL bronze.products"))
display(spark.sql("SELECT * FROM bronze.customers"))
display(spark.sql("SELECT * FROM bronze.orders"))
display(spark.sql("SELECT * FROM bronze.order_items"))
display(spark.sql("SELECT * FROM bronze.products"))

```
